```python
date = ['7/15/2020', '7,18,2020', '7,15,2020', '7,20/2020', '7,27,2020', '7,30,2020', '7,11,2020']
count_mask = [8, 11, 19, 15, 18, 6, 9]
count_nomask = [22, 19, 28, 29, 30, 21, 23]

DF = pd.DataFrame({'date': date, 'cases_mask': count_mask, 'cases_nomask': count_nomask})
DF['date'] = pd.to_datetime(DF['date'].str.replace(',', '/'), errors='coerce')
DF = DF.sort_values('date')

plt.figure(figsize=(10, 6))
plt.plot(DF['date'], DF['cases_mask'], marker='o', label='COVID Cases (Mask)', color='blue')
plt.plot(DF['date'], DF['cases_nomask'], marker='o', label='COVID Cases (No Mask)', color='red')
plt.xlabel('Date')
plt.ylabel('Number of COVID Cases')
plt.title('COVID Cases: Mask vs No Mask Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
